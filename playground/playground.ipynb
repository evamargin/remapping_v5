```python
import re
import numpy as np
import pandas as pd

_MARKER_PATTERNS = [
    re.compile(r'^(.+?):Marker\s+(\d+)$'),  # v1.25: 'BOF:Marker 0010'
    re.compile(r'^(.+)_(\d+)$'),            # v1.21: 'Rat_NP_3'
]

def _parse_marker_name(name):
    for pat in _MARKER_PATTERNS:
        m = pat.match(name)
        if m:
            return m.group(1), int(m.group(2))   # int() also fixes the '0010' corruption
    return None, None

def _locate_rows(dff):
    """Find metadata row indices by content, so both versions work."""
    group_row = None
    for i in range(4, 8):                         # group row is 4 (v1.21) or 5 (v1.25)
        cells = dff.iloc[i].fillna('').astype(str).str.lower()
        if cells.str.contains('rotation|position').any():
            group_row = i
            break
    if group_row is None:
        raise ValueError("Could not locate Rotation/Position row — unexpected format")
    return {
        'type': 1, 'name': 2,
        'parent': 4 if group_row == 5 else None,  # present only in newer format
        'group': group_row,
        'axis': group_row + 1,
        'data_start': group_row + 2,
    }

def _print_tree(d, indent='', show_leaves=True):
    keys = list(d.keys())
    for i, k in enumerate(keys):
        last   = (i == len(keys) - 1)
        branch = '└── ' if last else '├── '
        v = d[k]
        if isinstance(v, dict):
            print(f'{indent}{branch}{k}')
            _print_tree(v, indent + ('    ' if last else '│   '), show_leaves)
        elif show_leaves:
            if isinstance(v, np.ndarray):
                desc = f'array {v.shape} {v.dtype}'
            elif isinstance(v, (str, int, float, type(None))):
                desc = repr(v)
            else:
                desc = type(v).__name__
            print(f'{indent}{branch}{k}: {desc}')

def parse_motive(dff, verbose=False):
    rows = _locate_rows(dff)

    types   = dff.iloc[rows['type']].fillna('').astype(str).tolist()
    names   = dff.iloc[rows['name']].fillna('').astype(str).tolist()
    groups  = dff.iloc[rows['group']].fillna('').astype(str).tolist()
    axes    = dff.iloc[rows['axis']].fillna('').astype(str).tolist()
    parents = (dff.iloc[rows['parent']].fillna('').astype(str).tolist()
               if rows['parent'] is not None else [''] * dff.shape[1])

    data = dff.iloc[rows['data_start']:].apply(pd.to_numeric, errors='coerce')

    result = {'time_s': None, 'frames': None, 'rigid_body': {}, 'markers': {}}

    def rb_slot(n):     return result['rigid_body'].setdefault(n, {'rotation': {}, 'position': {}})
    def marker_slot(n, parent): return result['markers'].setdefault(n, {'position': {}, 'parent': parent})

    for col in range(dff.shape[1]):
        kind   = types[col].strip()
        name   = names[col].strip()
        parent = parents[col].strip()
        group  = groups[col].strip().lower()
        axis   = axes[col].strip().lower()
        values = data.iloc[:, col].to_numpy()

        # frame / time live in the axis row in both versions
        if axis == 'frame':
            result['frames'] = values.astype(np.int64); continue
        if axis.startswith('time'):
            result['time_s'] = values; continue

        if not name or not axis:
            continue

        if kind == 'Rigid Body' and group in ('rotation', 'position'):
            rb_slot(name)[group][axis] = values
            continue

        if kind in ('Rigid Body Marker', 'Marker') and group == 'position':
            body_from_name, num = _parse_marker_name(name)
            if body_from_name is None:
                if verbose: print(f'col {col}: bad marker name {name!r}')
                continue
            body = parent or body_from_name      # explicit Parent wins; else derive from name
            marker_slot(f'{body}_{num}', parent=body)['position'][axis] = values
            continue

        # 'Marker Quality' (v1.21) and anything else falls through here
        if verbose:
            print(f'col {col}: skipped kind={kind!r} group={group!r}')
            _print_tree(result)

    return result
```

```python
parsed_dicts = []
for table in ddf:
    parsed_dicts.append(preprocessing.parse_motive(table))
```
